** Standardize the variable/features to get your data ready for knn**<br>
** import `StandardScaler`**

```python
from sklearn.preprocessing import StandardScaler
```

**Create object of `scaler`**

```python
scaler = StandardScaler()
```

**Split the data into `features` and `target` to fit the `scaler` to the features only!**

```python
target = df['diagnosis']
features = df.drop('diagnosis',axis=1)
```

```python
features
```

```python
target
```

** Fit `scaler` to the features now!**

```python
scaler.fit(features)
```

** Please get the scaled features into `scaled_features` **

```python
scaled_features = scaler.transform(features)
scaled_features
```

```python
cols = df.columns[:-1]
cols
```

```python
df_scaled_features = pd.DataFrame(data=scaled_features, columns=cols)
df_scaled_features
```

** Do the train_test split** <br>Please use `test_size=0.33, random_state=42`

```python
from sklearn.model_selection import train_test_split
```

```python
X = df_scaled_features
y = target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=2)  
```

```python
X_train
```

```python
y_train
```

**import the `KNeighborsClassifier`**

```python
from sklearn.neighbors import KNeighborsClassifier
```

** Create a KNN model instance with n_neighbors=1**

```python
knn = KNeighborsClassifier(n_neighbors=1) 
```

**Fit the model to the training data**

```python
knn.fit(X_train, y_train)
```

**Do the predictions for the test data**

```python
predictions = knn.predict(X_test)
predictions
```

```python
knn.score(X_test, y_test)
```

```python
knn.predict([[17.99,10.38,122.80,1001.0,0.11840,0.27760,0.30010,0.14710,0.2419,0.07871]])
```

** Print confusion matrix and classification report**

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,predictions))
```
