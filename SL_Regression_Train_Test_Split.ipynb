```python
new_df = df.drop('per_capita_income_USD',axis='columns')
new_df
```

```python
income = df.per_capita_income_USD
income
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(new_df, income)
```

```python
reg.predict([[2020]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
(2020  * 828.46507522) + (-1632210.7578554575)
```

```python
bestfit_point = df.year*reg.coef_ + reg.intercept_
bestfit_point
plt.xlabel('year')
plt.ylabel('per capita income')

plt.scatter(df.year,df.per_capita_income_USD,color='blue',marker='+')
plt.plot(df.year,bestfit_point,color='red',marker='+')
```
